```python
import pandas as pd
import os
import numpy as np
```

```python
os.getcwd()
```

```python
home_data = pd.read_csv('train.csv',index_col = 'Id')
test = pd.read_csv('test.csv',index_col = 'Id')
```

```python
home_data.head()
```

```python
# find the missing values
missing_values = home_data.isnull().sum()
```

```python
missing_values  # gives all the columns with missing or nan values count
```

```python
type(missing_values)
```

```python
missing_values[missing_values>0]  # applying the condition to get only the missing values columns
```

```python
home_data.shape
```

```python
# the target which is to be predicted ---> remove the missing or nan values
# by considering whole dataframe
home_data.dropna(axis=0,subset=['SalePrice'],inplace=True)
```

```python
home_data.shape
```

```python
# creating or getting target from predictor
y = home_data.SalePrice   # series
```

```python
# removing the target from predictor
home_data.drop(['SalePrice'],axis=1,inplace=True)
```

```python
home_data.shape
```

```python
# to keep things simple we will use only numerical predictors
X = home_data.select_dtypes(exclude=['object'])  # DataFrame
X_test = test.select_dtypes(exclude=['object'])
```

```python
X.shape
```

```python
# Break off validation set from train
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
print(X_train.shape)
```

```python
# now we should remove the nan values from the columns 
# for that we must know the columns which have the nan or missing values

missing_values_in_columns = X_train.isnull().sum()
missing_values_in_columns[missing_values_in_columns>0]
```

```python
# these many columns have the nan or missing values 
lst = [i for i in X_train.columns if X_train[i].isnull().any()]
lst
```

```python
# now remove the nan or missing values from the X_train and X_valid dataframe
reduced_X_train = X_train.drop(lst,axis=1)
reduced_X_valid = X_valid.drop(lst,axis=1)
```

```python
reduced_X_train.head()
```

```python
reduced_X_train.shape
```

```python
# Train --> Predict ---> calculate mean absolute error

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def check(X_train,X_valid,y_train,y_valid):
    model = RandomForestRegressor(n_estimators=100,random_state=0)
    model.fit(X_train,y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid,preds)
```

```python
print(check(reduced_X_train,reduced_X_valid,y_train,y_valid))
```

```python
# the above result is for removing the nan or missing values
# now checking with some other approach to minimize or reduce the MAE
# so this time removing the missing values with the mean value (Imputation)
```

```python
# Imputation

from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
# this SimpleImputer is used to fill the missing or nan values 
# with either 'mean' or 'median'
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
```

```python
imputed_X_train.head()
```

```python
# as we can see above the imputation remove the columns names
# to put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```python
# see that our imputed_X_train have columns now
imputed_X_train.head()
```

```python
print('checking with other approach, MAE is : ')
print(check(imputed_X_train,imputed_X_valid,y_train,y_valid))
```

```python
'''Filling the missing values with 'mean' is considered good instead 
of removing the columns which are having some missing values.
But as we can see from our two approached that 2nd approach is 
is not good as per the dataset given'''
```

```python
# now considering one more approach(3rd) ---> 'median'
# preprocessed training and validation features
final_imputer = SimpleImputer(strategy='median')
final_X_train = pd.DataFrame(final_imputer.fit_transform(X_train))
final_X_valid = pd.DataFrame(final_imputer.transform(X_valid))
# imputation removes columns name, put them back
final_X_train.columns = X_train.columns
final_X_valid.columns = X_valid.columns
```

```python
print('check with 3rd approach, MAE is : ')
print(check(final_X_train,final_X_valid,y_train,y_valid))
```

```python
'''By comparing all approaches it is seen that our 3rd approach('median') 
is good as it gives less mean absolute error'''
```

```python
# now we will make the separate model for the better approach
# because we will soon use the trained model to generate test predictions

# Define and fit model
model = RandomForestRegressor(n_estimators = 100, random_state=0)
model.fit(final_X_train,y_train)

# get validation predictions and MAE
preds_valid = model.predict(final_X_valid)

# checking the mean absolute error
print('MAE is : ')
print(mean_absolute_error(y_valid,preds_valid))
```

```python
# now analysing our test data for prediction 
X_test.head()
```

firstly set the preprocessed test features to final_X_test.
